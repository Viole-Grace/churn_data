```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Churn_Modelling.csv")
df.shape
df.columns
df.info
```

```python
df.drop(['RowNumber','CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
X=df.iloc[: , :11].values
print(X[0])
Y=df.iloc[: ,-1].values
print(Y)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

le_c1=LabelEncoder()
le_c2=LabelEncoder()

X[:,1] = le_c1.fit_transform(X[:,1]) #fit for all values in column 1
X[:,2] = le_c2.fit_transform(X[:,2]) #fit for all values in column 2

ohe_c1 = OneHotEncoder(categorical_features=[1]) #the categorical features we want to OHE are in column 1
X=ohe_c1.fit_transform(X).toarray()
X=X[:, 1:] #remove one column to avoid dummy variable trap
```

```python
from sklearn.model_selection import train_test_split
x_tr, x_te, y_tr, y_te = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_tr=sc.fit_transform(x_tr)
x_te=sc.fit_transform(x_te)
print(x_te.shape)
print(x_te[0])
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()
model.add(Dense(output_dim=6, input_dim=12, activation='relu', init='uniform')) #input
model.add(Dense(output_dim=6, activation='relu', init='uniform')) #hidden layer 1
model.add(Dense(output_dim=1, activation='sigmoid', init='uniform')) #output
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_tr,y_tr, batch_size=10, nb_epoch=50)
```

```python
y_pred = model.predict(x_te)
for item in y_pred:
    print("%0.2f"%item)
y_pred = (y_pred > 0.5)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_te,y_pred)
print(cm)
```

```python
diag_sum,other_diag=0,0
for i in range(2):
    for j in range(2):
        if i==j:
            diag_sum+=cm[i,j]
        elif i!=0 and j!=0 and i%j==1:
            other_diag+=cm[i,j]
denom=diag_sum+other_diag
print('Accuracy : ',diag_sum/denom *100,'%')
```
